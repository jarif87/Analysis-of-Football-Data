```python
df['lwb'].fillna("0+0", inplace=True)
```

```python
df["lwb"] = df['lwb'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))
```

```python
df['ldm'].fillna("0+0", inplace=True)
df["ldm"] = df['ldm'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))

df['cdm'].fillna("0+0", inplace=True)
df["cdm"] = df['cdm'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))

df['rdm'].fillna("0+0", inplace=True)
df["rdm"] = df['rdm'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))

df['rwb'].fillna("0+0", inplace=True)
df["rwb"] = df['rwb'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))

df['lb'].fillna("0+0", inplace=True)
df["lb"] = df['lb'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))

df['lcb'].fillna("0+0", inplace=True)
df["lcb"] = df['lcb'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))


df['cb'].fillna("0+0", inplace=True)
df["cb"] = df['cb'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))


df['rcb'].fillna("0+0", inplace=True)
df["rcb"] = df['rcb'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))


df['rb'].fillna("0+0", inplace=True)
df["rb"] = df['rb'].str.split('+').apply(lambda x: int(x[0]) + int(x[1]))
```

```python
df= df[["age", "height_cm", "weight_kg","overall", "potential", "value_eur", "wage_eur", "player_positions", "preferred_foot", "lwb", "ldm", "cdm", "rdm", "rwb", "lb", "lcb", "cb", "rcb", "rb"]]
```

```python
df['player_positions'] = df['player_positions'].str.split(',').str[0]
```

```python
df['player_positions'].value_counts()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(y="player_positions",data=df,palette="Set1")
plt.show()
```

```python
replacement_dict = {'CB': 0, 'ST': 1, 'CM': 2, 'GK': 3, 'CDM': 4, 'RB': 5, 
                    'LB': 6, 'CAM': 7, 'RM': 8, 'LM': 9, 'LW': 10, 'RW': 11, 
                    'CF': 12, 'LWB': 13, 'RWB': 14}

df['player_positions'] = df['player_positions'].replace(replacement_dict)
```

```python
r={"Right":0,"Left":1}
df["preferred_foot"]=df["preferred_foot"].replace(r)
```

```python
columns_to_process = ['lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb']


for col in columns_to_process:
    df[col].replace(0, df[col].mean(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler


columns_to_scale = ['age', 'height_cm', 'weight_kg', 'overall', 'potential', 'value_eur', 'wage_eur',
                    'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb']
scaler = StandardScaler()

df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df.head()
```

```python

train=df.drop(columns=["player_positions"],axis=1)
label=df["player_positions"]
```

```python
from imblearn.combine import SMOTETomek
smote_tomek = SMOTETomek(random_state=42)
X,Y = smote_tomek.fit_resample(train,label)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

# RandomForest

```python

rf = RandomForestClassifier(n_estimators=500, random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
a= {'CB': 0, 'ST': 1, 'CM': 2, 'GK': 3, 'CDM': 4, 'RB': 5, 
                    'LB': 6, 'CAM': 7, 'RM': 8, 'LM': 9, 'LW': 10, 'RW': 11, 
                    'CF': 12, 'LWB': 13, 'RWB': 14}
label_name=list(a.keys())
```

```python
report = classification_report(y_test, y_pred,target_names=label_name)
print(report)
```

```python
plt.figure(figsize=(10,6))
cm= confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True,fmt="d",xticklabels=label_name,yticklabels=label_name,cmap="rainbow",lw=4,linecolor="navy")
plt.title("Confusion Matrix")
plt.show()
```

```python
from yellowbrick.classifier import ROCAUC
cl_name=['CB','ST','CM','GK','CDM','RB','LB','CAM','RM','LM','LW','RW','CF','LWB','RWB']
model = rf
visualizer = ROCAUC(model, classes=cl_name)

visualizer.fit(X_train, y_train)        # Fit the training data to the visualizer
visualizer.score(X_test, y_test)        # Evaluate the model on the test data
visualizer.show() 
```

# CatBooster Classifier

```python
from catboost import CatBoostClassifier

cat = CatBoostClassifier(iterations=1000, depth=6, learning_rate=0.1, loss_function='MultiClass', verbose=200)

cat.fit(X_train, y_train, eval_set=(X_test, y_test))
```

```python
cat_pred = cat.predict(X_test)
accuracy0 = accuracy_score(y_test, cat_pred)
print(f'Accuracy: {accuracy0*100:.2f}%')
```

```python
print(classification_report(y_test, cat_pred,target_names=label_name))
```

```python
plt.figure(figsize=(10,6))
cm= confusion_matrix(y_test, cat_pred)
sns.heatmap(cm,annot=True,fmt="d",xticklabels=label_name,yticklabels=label_name,cmap="tab10",lw=4,linecolor="red")
plt.title("Confusion Matrix")
plt.show()
```

```python

from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
import numpy as np
y_pred = cat.predict_proba(X_test)

# Binarize the labels
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_bin.shape[1]

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_pred.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
plt.figure(figsize=(8, 6))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,label=f'ROC curve (class {i}) (AUC = {roc_auc[i]:.2f})')

plt.plot(fpr["micro"], tpr["micro"],color='deeppink', lw=2, linestyle='--',label='Micro-average ROC curve (AUC = {0:.2f})'.format(roc_auc["micro"]))

plt.plot([0, 1], [0, 1], 'k--', lw=4)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
